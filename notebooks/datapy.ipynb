#### MNIST Data Loader (`data.py`) – Exploration and Verification

This notebook documents and tests the `load_mnist()` function defined in `src/data.py`.

Goals:

- Verify that the MNIST IDX files are correctly loaded and decompressed.
- Confirm shapes, dtypes, and value ranges of train/test splits.
- Visualize a few images to ensure the flattening/normalization is correct.
- Check that the API of `load_mnist()` is stable and convenient for `train.py` and `model.py`.

```python
from pathlib import Path
import sys

import numpy as np
import matplotlib.pyplot as plt

project_root = Path.cwd()
src_path = project_root / "src"

if str(src_path) not in sys.path:
    sys.path.append(str(src_path))

from data import load_mnist
```

#### 1. Load MNIST via `load_mnist()`

Expect the loader to:

- Read the 4 IDX `.gz` files from `data/mnist` (or similar).
- Normalize pixel values to `[0, 1]` as float32.
- Flatten 28×28 images to vectors of length 784.
- Split into `(X_train, y_train, X_test, y_test)`. 

```python
X_train, y_train, X_test, y_test = load_mnist()

print("X_train:", X_train.shape, X_train.dtype, X_train.min(), X_train.max())
print("y_train:", y_train.shape, y_train.dtype)
print("X_test:", X_test.shape, X_test.dtype, X_test.min(), X_test.max())
print("y_test:", y_test.shape, y_test.dtype)
```

##### 2. Visual sanity check

We reshape a few `X_train` rows back to 28×28 and plot them with their labels.

This verifies that:

- The flattening order matches the IDX format.
- Normalization preserves the digit structure. 

```python
def show_example(idx: int) -> None:
    img = X_train[idx].reshape(28, 28)
    label = y_train[idx]
    plt.imshow(img, cmap="gray")
    plt.title(f"Index {idx}, Label {label}")
    plt.axis("off")
    plt.show()

for i in range(5):
    show_example(i)
```

##### 3. API contract for `load_mnist()`

For the rest of the project we rely on:

- `X_train`: shape `(N_train, 784)`, dtype `float32`, values in `[0, 1]`.
- `y_train`: shape `(N_train,)`, dtype integer labels `0–9`.
- `X_test`: shape `(N_test, 784)`, same preprocessing as `X_train`.
- `y_test`: shape `(N_test,)`.

If the loader implementation changes (e.g. different shapes, different normalization),
this notebook should be updated and the contract re-verified.
